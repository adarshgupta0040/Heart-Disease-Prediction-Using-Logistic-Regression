```python
X = df.drop('HeartRisk',axis = 1)
X=X.drop("P_ID",axis=1)
y = df["HeartRisk"]
```

```python
X = X.apply(lambda x: x.fillna(x.mean()),axis=0)
X.isnull().sum(axis = 0)
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression()
LR.fit(X_train, y_train)
```

```python
y_pred = LR.predict(X_test)
```

```python
print('Train Score is : ' , LR.score(X_train, y_train))
print('Test Score is : ' , LR.score(X_test, y_test))
```

```python
from sklearn.metrics import classification_report
print('Classification Report is: \n \n' , classification_report(y_test, y_pred ))
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

print('The accuracy score is:', accuracy_score(y_test, y_pred))
```
